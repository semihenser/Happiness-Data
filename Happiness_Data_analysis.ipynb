## Top 10 countries for each attribute

```python
fig, axes = plt.subplots(nrows=4, ncols=2,constrained_layout=True,figsize=(12,8))

sns.barplot(x='Logged GDP per capita',y='Country name',data=data.nlargest(10,'Logged GDP per capita'),ax=axes[0,0],palette="Blues_d")

sns.barplot(x='Social support' ,y='Country name',data=data.nlargest(10,'Social support'),ax=axes[0,1],palette="YlGn")

sns.barplot(x='Healthy life expectancy' ,y='Country name',data=data.nlargest(10,'Healthy life expectancy'),ax=axes[1,0],palette='OrRd')

sns.barplot(x='Freedom to make life choices' ,y='Country name',data=data.nlargest(10,'Freedom to make life choices'),ax=axes[1,1],palette='YlOrBr')

sns.barplot(x='Generosity' ,y='Country name',data=data.nlargest(10,'Generosity'),ax=axes[2,0],palette='Blues_d')

sns.barplot(x='Ladder score' ,y='Country name',data=data.nlargest(10,'Ladder score'),ax=axes[2,1],palette='YlGn')

sns.barplot(x='Ladder score' ,y='Regional indicator',data=data.nlargest(10,'Ladder score'),ax=axes[3,0],palette='OrRd')

sns.barplot(x='Generosity' ,y='Regional indicator',data=data.nlargest(10,'Generosity'),ax=axes[3,1],palette='OrRd')
```

## Worst 10 countries for each attribute

```python
fig, axes = plt.subplots(nrows=3, ncols=2,constrained_layout=True,figsize=(12,8))

sns.barplot(x='Logged GDP per capita',y='Country name',data=data.nsmallest(10,'Logged GDP per capita'),ax=axes[0,0],palette="Blues_d")

sns.barplot(x='Social support' ,y='Country name',data=data.nsmallest(10,'Social support'),ax=axes[0,1],palette="YlGn")

sns.barplot(x='Healthy life expectancy' ,y='Country name',data=data.nsmallest(10,'Healthy life expectancy'),ax=axes[1,0],palette='OrRd')

sns.barplot(x='Freedom to make life choices' ,y='Country name',data=data.nsmallest(10,'Freedom to make life choices'),ax=axes[1,1],palette='YlOrBr')

sns.barplot(x='Generosity' ,y='Country name',data=data.nsmallest(10,'Generosity'),ax=axes[2,0],palette='Blues_d')

sns.barplot(x='Ladder score' ,y='Country name',data=data.nsmallest(10,'Ladder score'),ax=axes[2,1],palette='YlGn')
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot

data = dict(type = 'choropleth', 
          locations = data['Country name'],
           locationmode = 'country names',
           colorscale='RdYlGn',
           z = data['Ladder score'], 
           text = data['Country name'],
           colorbar = {'title':'Ladder score'})

layout = dict (title = 'Geographical Visualization of Happiness Score', 
              geo = dict(showframe = True, projection = {'type': 'azimuthal equal area'}))

choromap3 = go.Figure(data = [data], layout=layout)
iplot(choromap3)
```
